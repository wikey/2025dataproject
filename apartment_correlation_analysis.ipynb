```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('/mnt/data/df_apartment_filtered.csv')

# Select variables
selected = ['price','minimum_nights','number_of_reviews','reviews_per_month','availability_365']
df_filtered = df[selected]

df_filtered.head()
```

```python
corr = df_filtered.corr(numeric_only=True)
corr
```

```python
plt.figure(figsize=(7,6))
plt.imshow(corr, aspect='auto')
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap â€“ Selected Apartment Variables")
plt.tight_layout()
plt.show()
```
