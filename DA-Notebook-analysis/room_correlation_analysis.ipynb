# Room Listings Correlation Analysis

## 1. Import Libraries

```
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. Load Room Dataset

```
df_room = pd.read_csv('/mnt/data/df_room_filtered.csv')
df_room.head()
```

## 3. Select Variables

```
selected = ['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365']
df_filtered_room = df_room[selected]
df_filtered_room.head()
```

## 4. Compute Correlation Matrix

```
corr_room = df_filtered_room.corr(numeric_only=True)
corr_room
```

## 5. Heatmap Visualization

```
plt.figure(figsize=(7, 6))
plt.imshow(corr_room, aspect='auto')
plt.colorbar()
plt.xticks(range(len(corr_room.columns)), corr_room.columns, rotation=90)
plt.yticks(range(len(corr_room.columns)), corr_room.columns)
plt.title("Correlation Heatmap â€“ Selected Room Listing Variables")
plt.tight_layout()
plt.show()
```
